## B. Data Analysis with Time Series data

### 1. Read dataset

```python
dataset_path = "data/opsd_germany_daily.csv"

opsd_daily = pd.read_csv(dataset_path)

print(opsd_daily.shape)
print(opsd_daily.dtypes)
opsd_daily.head()
```

For Time Series data, we can set the Date column as the index.

```python
opsd_daily = opsd_daily.set_index("Date")
opsd_daily.head(3)
```

```python
opsd_daily = pd.read_csv(dataset_path, index_col=0, parse_dates=True)

opsd_daily["Year"] = opsd_daily.index.year
opsd_daily["Month"] = opsd_daily.index.month
opsd_daily["Weekday Name"] = opsd_daily.index.day_name()

opsd_daily.sample(5, random_state=0)
```

### 2. Time-based indexing

Select the data samples from 2014-01-20 to 2024-01-22

```python
opsd_daily.loc["2014-01-20":"2024-01-22"]
```

`Partial-string indexing` allows you to retrieve all data samples belonging to '2012-02' by simply specifying '2012-02'.

```python
opsd_daily.loc["2012-02"]
```

### 3. Visualizing time series data

The graph shows the daily electricity consumption in Germany

```python
# Use seaborn style defaults and set the defalut figure size
sns.set(rc={"figure.figsize":(11 , 4)})
opsd_daily["Consumption"].plot(linewidth=0.5)
```

The graph shows electricity consumption, solar energy production, and wind energy production.

```python
cols_plot = ["Consumption", "Solar", "Wind"]
axes = opsd_daily[cols_plot].plot(marker=".", alpha=0.5, linestyle="None", figsize=(11, 9), subplots=True)

for ax in axes:
    ax.set_ylabel("Daily Total (GWn)")

plt.show()
```

### 4. Seasonality

Visualize the distribution of the columns Consumption, Wind, and Solar by month.

```python
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)

for name, ax in zip(["Consumption", "Solar", "Wind"], axes):
    sns.boxplot(data=opsd_daily, x="Month", y=name, ax=ax)
    ax.set_ylabel("GWh")
    ax.set_title(name)
    # Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel("")
```

### 5. Frequencies

Get frequency by day from 1998-03-10 to 1998-03-15

```python
pd.date_range("1998-03-10", "1998-03-15", freq="D")
```

Use `forward fill` to replace missing data by using the last recorded value until a new one is encountered.

```python
times_sample = pd.to_datetime(["2013-02-03", "2013-02-06", "2013-02-08"])
consum_sample = opsd_daily.loc[times_sample, ["Consumption"]].copy()
consum_sample
```

```python
# Convert the data to daily frequency, without filling any missings
consum_freg = consum_sample.asfreq("D")

# Create a column with missings forward filled
consum_freg["Consumption - Forward Fill"] = consum_sample.asfreq("D", method="ffill")
consum_freg
```

### 6. Resampling


Downsample from daily to weekly frequency, where the column values become the 7-day average for each week.

```python
data_columns = ["Consumption", "Wind", "Solar", "Wind+Solar"]

opsd_weekly_mean = opsd_daily[data_columns].resample("W").mean()
opsd_weekly_mean.head(3)
```

```python
print(opsd_weekly_mean.shape)
print(opsd_weekly_mean.shape)
```

Visualize both daily and weekly time series of solar data over 6 months

```python
start, end = "2017-01", "2017-06"

fig, ax = plt.subplots()
ax.plot(
    opsd_daily.loc[start:end, "Solar"],
    marker=".",
    linestyle="-",
    linewidth=0.5,
    label="Daily"
)
ax.plot(
    opsd_weekly_mean.loc[start:end, "Solar"],
    marker="o",
    markersize=8,
    linestyle="-",
    label="Weekly Mean Resample"
)
ax.set_ylabel("Solar Production (GWh)")
ax.legend()
plt.show()
```

min_count=360 means that to compute the annual sum, there must be at least 360 valid (non-missing) values in the year. If a year has fewer than 360 valid values, the result will be NaN instead of a total.

```python
opsd_annual = opsd_daily[data_columns].resample("Y").sum(min_count=360)

opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual.index.name = "Year"
opsd_annual["Wind+Solar/Consumption"] = opsd_annual["Wind+Solar"] / opsd_annual["Consumption"]
opsd_annual.tail(3)
```

Chart showing the contribution of wind and solar energy production to electricity consumption since 2012

```python
ax = opsd_annual.loc[2012:, "Wind+Solar/Consumption"].plot.bar(color="C0")
ax.set_ylabel("Fraction")
ax.set_ylim(0, 0.3)
ax.set_title("Wind + Solar Share of Annual Electricity Consumption")
plt.xticks(rotation=0)
plt.show()
```

### 7. Rolling windows

```python
opsd_7d = opsd_daily[data_columns].rolling(7, center=True).mean()
opsd_7d.head(10)
```

### 8. Trends

```python
import matplotlib.dates as mdates

opsd_365d = opsd_daily[data_columns].rolling(window=365, center=True, min_periods=360).mean()

fig, ax = plt.subplots()
ax.plot(
    opsd_daily["Consumption"],
    marker=".",
    markersize=2,
    color="0.6",
    linestyle="None",
    label="Daily"
)
ax.plot(opsd_7d["Consumption"], linewidth=2, label="7-d Rolling Mean")
ax.plot(
    opsd_365d["Consumption"],
    color="0.2",
    linewidth=3,
    label="Trend (365-d Rolling Mean)"
)

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel("Year")
ax.set_ylabel("Consumption (GWh)")
ax.set_title("Trends in Electricity Consumption")
plt.show()
```

```python
fig, ax = plt.subplots()
for nm in ["Wind", "Solar", "Wind+Solar"]:
    ax.plot(opsd_365d[nm], label=nm)
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.set_ylim(0, 400)
    ax.legend()
    ax.set_ylabel("Production (GWh)")
    ax.set_title("Trends in Electricity Production (365-d Rolling Means)")
plt.show()
```
